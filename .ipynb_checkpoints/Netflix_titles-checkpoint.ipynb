```python
#Phase 1:- Data loading & understanding

#Task-1:Load dataset
import pandas as pd
df=pd.read_csv("netflix_titles.csv")
df2=df.copy()
```

```python
#Task-2:-
#Frist 5 rows
df2.head(5)
#Shape
df2.shape
#columns
df2.columns
#Data types
df2.dtypes
#Missing value
df2.duplicated()
```

```python
#Phase 2:- Data Cleaning

#Task-3:- Convert date_added into datatime format
df2["date_added"]=pd.to_datetime(df2["date_added"],errors="coerce")
#TASK 4:Remove duplicate rows
df2=df2.drop_duplicates()
# TASK 5:Clean string columns:country,director,cast,listed_in
df2[["country","director","cast","listed_in"]]=df2[["country","director","cast","listed_in"]].apply(lambda col:col.str.strip())
# TASK 6:Handle missing values properly
df2[["director","cast","country"]]=df2[["director","cast","country"]].apply(lambda i:i.fillna("Unkown"))
df2
```

```python
#Phase 3:- Basic Analysis
#Task 7:- Count:- Total Movie & Total Tv Shows
df2["type"].value_counts()
#Task 8:- Find:- Most common rating
df2["rating"].value_counts().head()
#Task 9:- Find:- Top 10 countries producing Netflix content
df2["country"].str.split(', ').explode().value_counts().head(10)
#Task 10:- Find:- Top 10 genres(listed_in)
df2["listed_in"].str.split(', ').explode().value_counts().head(10)
#Task 11:- Find:- Top 10 directors
df2["director"].str.split(', ').explode().value_counts().head(10)
```

```python
#Phase 4: Date & Trend Analysis
#Task 12:- Which year had maximum content added to netflix
df2["date_added"].dt.year.value_counts().head(1)
#Task 13:- Show Netflix growth year by year
df2["date_added"].dt.year.value_counts().sort_index()
#Task 14:- Find oldest releae year in dataset
df2["date_added"].dt.year.value_counts().sort_index().head(1)
#Task 15:- Find newest release year in datset
df2["date_added"].dt.year.value_counts().sort_index(ascending=False).head(1)
```

```python
#Phase 5: Duration Analysis
#Task 16:- Find average movie duration
Movie=df2[df2["type"]=="Movie"]
Movie_duration=Movie["duration"].str.extract('(\d+)')[0].astype(float).mean()
Movie_max_duration=Movie["duration"].str.extract('(\d+)')[0].astype(float).max()
Movie_min_duration=Movie["duration"].str.extract('(\d+)')[0].astype(float).min()

print(Movie_duration,'\n',Movie_max_duration,'\n',Movie_min_duration)
```

```python
#Phase 6:- Filtering Practice
#Task 19: Fing :- Indian movie after 2018
Indian_Movie=df2[(df2["type"]=="Movie") & (df2["release_year"]>2018)]
Indian_Movie_TV_MA_rating=Indian_Movie[(Indian_Movie["rating"]=="TV-MA")&(Indian_Movie["duration"].str.extract('(\d+)')[0].astype(float)>120)]
Rajiv_Chilaka=df2[df2["director"]=="Rajiv Chilaka"]
cast_salman=df2[df2["cast"].str.contains("salman khan",case=False,na=False)]
cast_salman
```

```python
#Phase 7:- GroupBy
#Task 23:- Count content by release year
df2.groupby("release_year").size()
#Task 24:- count content by rating
df2.groupby("rating").size()
#Task 25:- Count movie and Tv shows by country
df2["country"].dropna().str.split(', ').explode().str.strip().groupby(df2["type"]).size()
```

```python
#Phase 8:- Visualization 
#Task 26:-Bar chart:- Movies VS TV Shows
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
df2["type"].value_counts().plot(kind="bar")
plt.title("Movies VS TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()

#Task 27:- Line chart:- Content added per year
plt.figure(figsize=(8,6))
df2["release_year"].value_counts().sort_index().plot(kind="line")
plt.title("Content Added Per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()

#Task 28:- Bar chart:- Top 10 countries
plt.figure(figsize=(8,5))
df2["country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Countries")
plt.xlabel("Coutries")
plt.ylabel("Count")
plt.show()

#Task 29:- Bar chart:- Top genres
plt.figure(figsize=(8,5))
df2["listed_in"].dropna().str.split(', ').explode().value_counts().head(10).plot(kind="bar")
plt.show()
```

```python
#Task 30 :- which month has most contet added?
df2["date_added"].dt.month_name().value_counts()
#Task 31:- Does Netflix focus more on Movies or TV shows over time
plt.figure(figsize=(8,5))
df2.groupby(["type","country"]).size().plot(kind="line")
plt.tight_layout()
#Task 32:- Compare India VS US content count
df2["country"].str.split(', ').explode().str.strip().value_counts()
#Task 33:- Which actor appears most on netflix
df2["cast"].str.split(', ').explode().str.strip().value_counts().head(1)
```
